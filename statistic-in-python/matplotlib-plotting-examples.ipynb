```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# y = x^2
x = np.array([1,2,3,4,5,6,7])
y = x**2
```

```python
%matplotlib inline
scale = [0, x.max]
plt.plot(x, y, 'ro')
plt.axis([0,x.max(),0,y.max()])  #takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes
```

```python
# red dashes, blue squares and green triangles
plt.plot(x, y, 'r--', x, x**3, 'bs', x, x**4, 'g^')

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fg = plt.figure(1)  #optional because figure(1) will be created by default
plt.subplot(211) #produces a subaxes in a figure which represents the top plot (i.e. the first) in a 2 row by 1 column
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.subplot(211)
plt.show()
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

## Work with text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.figure(1)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
#plt.show()

plt.figure(2)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.annotate('local max', xy=(104, 0.028), xytext=(130, 0.015),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.show()
```
